```python
import math

# Define the Haversine formula function
def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great-circle distance between two points on the Earth's surface.
    :param lat1: Latitude of point 1 (in decimal degrees)
    :param lon1: Longitude of point 1 (in decimal degrees)
    :param lat2: Latitude of point 2 (in decimal degrees)
    :param lon2: Longitude of point 2 (in decimal degrees)
    :return: Distance between the two points in kilometers
    """
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    # Radius of the Earth in kilometers (mean radius)
    r = 6371.0
    return c * r
```

```python
# Sample proximity points (these are locations near the launch sites)
proximity_data = {
    'Place': ['Nearby City 1', 'Nearby City 2'],
    'Lat': [28.4012, 34.5012],
    'Long': [-80.6100, -120.6020]
}

# Convert to DataFrame
proximity_df = pd.DataFrame(proximity_data)

# Loop through each launch site in spacex_df
for index, launch_site in spacex_df.iterrows():
    launch_site_name = launch_site['Launch Site']
    launch_site_lat = launch_site['Lat']
    launch_site_lon = launch_site['Long']
    
    # Calculate distances from this launch site to each proximity point
    proximity_df['distance_km'] = proximity_df.apply(
        lambda row: haversine(launch_site_lat, launch_site_lon, row['Lat'], row['Long']), axis=1
    )
    
    # Add markers for proximity points with calculated distance to the map
    for i, proximity in proximity_df.iterrows():
        # Create a popup showing the place name and distance from the launch site
        popup_text = f"{proximity['Place']} - {proximity['distance_km']:.2f} km from {launch_site_name}"
        
        # Add a marker for each proximity point
        folium.Marker(
            [proximity['Lat'], proximity['Long']],
            popup=popup_text,
            icon=folium.Icon(color='blue')
        ).add_to(site_map)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
